# Convert json to gpx
This script is designed to convert the json files generated by Swiftsure Scrape.ipynb to gpx files

```python
import json
import gpxpy #pip install gpxpy
import gpxpy.gpx
from datetime import datetime
import os
```

```python
trackFolders = ['C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2015 Stage 1/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2015 Stage 2/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2016 Stage 1/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2016 Stage 2/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2017 Stage 1/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2017 Stage 2/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2018 Stage 1/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2018 Stage 2/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2019 Stage 1/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2019 Stage 2/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/Seventy48 2018/tracks/',
                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/Seventy48 2019/tracks/']

#trackFolders = ['C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2015 Stage 1/tracks/',
#                'C:/Users/benja/Documents/GitHub/R2AK/Code/Swiftsure Scrape/data/R2AK 2015 Stage 2/tracks/']

for race in trackFolders:
    gpx = gpxpy.gpx.GPX() #initialization
    racename = os.path.basename(os.path.dirname((os.path.dirname(race))))
    print('Processing:', racename)
    outfilePath = os.path.dirname(os.path.dirname(race)) + '/tracks.gpx'
    for file in os.listdir(race):
        filename = file[:-5]
        with open(race+file) as json_file:
            gpx_track = gpxpy.gpx.GPXTrack(name=filename, description=racename) 
            gpx_track_segment = gpxpy.gpx.GPXTrackSegment()
            data = json.load(json_file)
            for waypoint in data['payload']:
                tp = gpxpy.gpx.GPXTrackPoint(latitude = waypoint['lt'],
                                             longitude = waypoint['ln'],
                                             time = datetime.fromtimestamp(waypoint['ts'])
                                            )
                tp.course = waypoint['h']
                tp.speed = waypoint['s']
                tp.comment = waypoint['stat']
                tp.description = waypoint['dstat']
                gpx_track_segment.points.append(tp)
            gpx_track.segments.append(gpx_track_segment)
            gpx.tracks.append(gpx_track)

    with open(outfilePath, 'w', ) as gpx_file:
        gpx_file.write(gpx.to_xml(version='1.0'))
print('done')
```
